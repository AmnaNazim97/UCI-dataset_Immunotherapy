```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
seed = 7
np.random.seed(seed)
```

```python
#Uplaoding file
```

```python
dataframe = pd.read_csv("immune.csv" , header=None)
dataframe

```

```python
dataset = dataframe.values
print(dataset[0])
print(dataset.ndim)
print(dataset.shape)
print(type(dataset))
```

```python
#Split 
```

```python
x= dataset[1:, 0:7].astype(float)
y = dataset[1:,7]
print(x[0])
print(y)
print(x.shape)
print(y.shape)
```

```python
#Create baseline model
```

```python
from keras import optimizers
```

```python
def create_baseline():
  network = Sequential()
  network.add(Dense(8, activation='relu', input_shape=(60,)))
  network.add(Dense(1, activation='sigmoid'))
  network.compile(optimizer= optimizers.Adam(lr=0.001), loss='binary_crossentropy', metrics=['acc'])
  return network
```

```python
#Run the model
```

```python
estimator = KerasClassifier(build_fn = create_baseline , epochs = 10, batch_size = 5 , verbose = 0 )
kfold = StratifiedKFold(n_splits = 10 , shuffle = True , random_state = seed)
results = cross_val_score(estimator , x , y , cv = kfold)
print("Results: %2f%%(%2f%%)"%(results.mean()*100,results.std()*100))
```

```python
#Create a larger network
def large_network():
    network = Sequential()
    network.add(Dense(8, activation='relu', input_shape=(7,)))
    network.add(Dense(1, activation='sigmoid'))
    network.compile(optimizer=optimizers.Adam(lr=0.01), loss='binary_crossentropy' , metrics=['acc'])
    return network

estimator = KerasClassifier(build_fn = create_baseline , epochs = 7, batch_size = 5 , verbose = 1)
kfold = StratifiedKFold(n_splits = 10 , shuffle = True , random_state = seed)
results = cross_val_score(estimator , x , y , cv = kfold)
print("Results: %2f%%(%2f%%)"%(results.mean()*100,results.std()*100))
```

```python
complete_data= dataset.copy()
np.random.shuffle(complete_data)
data = dataset[1:,0:7].astype('float32')
labels = complete_data[1:,7]
print(data.dtype)
```

```python
train_data = x[:50]
train_labels = y[:50]
```

```python
test_data = data[50:]
test_labels = data[50:]
```

```python
print(train_data.shape)
print(test_data.shape)
print(train_labels.shape)
print(test_labels.shape)
print(train_labels[0])
```

```python
def build_model():
  network = Sequential()
  network.add(Dense(10, activation='relu',input_shape=(7,)))
  network.add(Dense(1, activation='sigmoid'))
  network.compile(optimizer = optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-8), loss='binary_crossentropy', metrics=["accuracy"])
  return network
```

```python
k=4
num_val_samples = len(test_data) // k
num_epochs = 100

for i in range(k):
  print('processing fold #', i)
  val_data = train_data[i * num_val_samples: (i + 1) * num_val_samples]
  val_labels = train_labels[i * num_val_samples: (i + 1) * num_val_samples]
  partial_train_data = np.concatenate(
      [train_data[:i * num_val_samples],train_data[(i + 1) * num_val_samples:]],axis=0)
  partial_train_labels = np.concatenate(
      [train_labels[:i * num_val_samples],
       train_labels[(i + 1) * num_val_samples:]],axis=0)
  model = build_model()
  model.fit(partial_train_data, partial_train_labels , epochs=num_epochs, batch_size=5 , verbose = 0)
  loss , acc = model.evaluate(val_data, val_labels , verbose=0)
  print(acc)
```
